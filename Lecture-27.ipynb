Denetimli Öğrenme: Makineye (yani algoritmaya) biz daha önce gerçekleşmiş olan olayları ve sonuçları girdi olarak veririz. Makine bu olaylarla sonuçlar arasında bağlantı kurar. Daha sonra biz yeni bir olayı makineye verdiğimizde onun sonucunu makineden kestirmesini isteriz.

Denetimsiz Öğrenme: Biz makineye yalnızca olayları veririz. Makine bunların arasındaki benzerliklerden ve farklılıklardan hareketle bizim istediğimiz sonuçları çıkartmaya çalışır. Örneğin biz makineye resimler verip bunların elma mı armut mu olduğunu söyleyelim. Ve bunu çok miktarda yapalım. Sonra ona bir elma resmi verdiğimizde o daha önceki deneyimlerden hareketle bunun elma olduğu sonucunu çıkartabilecektir. İşte bu denetimli öğrenmeye örnek. Şimdi biz makineye elma ve armut resimlerini verelim ama bunların ne olduğunu söylemeyelim. Ondan bu resimleri ortak özelliklerine göre iki gruba ayırmasını isteyelim. Bu da denetimsizye örnek.

Pekiştirmeli Öğrenme:  Pekiştirmeli öğrenmede ise tıpkı edimsel koşullanmada olduğu gibi hedefe yaklaşan durumlar ödüllendirilerek makinenin öğrenmesi sağlanmaktadır. Ödül (award), Ceza (penalty)

## İstatistikte Ölçek türleri 

Ölçülen ya da ölçülmüş olan değerlerin sınıflarına genel olarak "ölçek (scale)" denilmektedir. 

- Kategorik (Nominal) Ölçekler: Kümenin elemanları kategorik olgulardır. cinsiyet, renk, coğrafi bölge. kadınlarla erkekler arasında sigara içme miktarı arasından anlamlı bir fark olup olmadığı --> cinsiyet, renk tercihi

- Sırasal (Ordinal) Ölçekler: birer kategori belirtirler ancak bu kategoriler arasında büyüklük-küçüklük ilişkisi vardır. eğitim durumu --> ilköğretim, lise, üniversite

- Aralıklı (Interval) Ölçekler: sayısal bilgi içerirler. iki puan arasındaki fark aynı miktar uzaklığı ya da yakınlığı ifade eder. sınavlardan alınan puanlar böyle bir ölçek türündendir. sınavdan sıfır alınabilir. ancak bu sıfır o kişinin o konu hakkında hiçbir şey bilmediği anlamına gelmez. yani mutlak sıfır değildir. ya da örneğin "derece" bir aralıklı ölçeği belirtmektedir. 50 derece ile 40 derece arasındaki ısı farklı 40 derece ile 30 derece arasındaki fark kadardır ancak sıfır derece ısının olmadığı anlamına gelmez. 

- Oransal (Ratio) Ölçekler: sayısal bilgi içerirler. mutlak bir sıfır noktası vardır. uzunluk, kütle. örneğin kişinin yaşı da oransal bir ölçek.

### Olasılık Kavramı

--> bir olgunun gerçekleşme beklentisi. tura gelme olasılığı 0.5 - %50. "göreli sıklık (relative frequency)." 

tura gelme olasılığı / n --> 0.5. 1000000 atıldığında yazı ile tura gelme sayısı aynı olmaz. ancak tura gelme oranı gitgide 0.5'e yakınsar. o halde bir bir paranın atılması durumunda tura gelme olasılığının 0.5 olması demekle aslında bir limit durumunu kastetmik olmaktayız



```python
import random

def head_tail(n):
    head = 0
    for _ in range(n):
        head += random.randint(0, 1) # tail = 0, head = 1
    
    return head / n
```

```python
head_tail(1)
```

```python
head_tail(10)
```

```python
head_tail(100)
```

```python
head_tail(10_000)
```

```python
head_tail(100_000)
```

```python
head_tail(1_000_000)
```

para atım sayısı arttıkça tura gelme olasılığı da 0.5'e yakınsamaktadır. büyük sayılar yasası (law of large numbers).

## Veri Kümelerinin Kullanıma Hazır Hale Getirilmesi

ilk adım verilerin toplanmasıdır. anket, sensörler vs. sosyal medya uygulamalarında kişilerin paylaştığı şeyler. verileri resmi kurumlar tarafından da toplanıyor veya sunuluyor olabilir. Türkiye istatistik enstitüsü. google tarafından yönetilen kaggle (kaggle.com).

Veriler toplandıktan sonra onların "veri kümeleri (datasets)" biçiminde düzenlenmesi gerekmektedir. 

medeni durum    yaş     eğitim      maaş

veri tabloları genellikle dosyalarda saklanır. csv. 

1) verilerin temizlenmesi (data cleaning): veriler içerisinde geçersiz değerler söz konusu olabilir. nan/null, bozulmuş veri biçiminde, yinelenen veriler biçiminde karşımıza çıkabilir. aşırı uçtaki değerer.

2) özellik seçimi (feature selection): özellik seçimi veri tablosundaki sütunların gerekli olanlarının alınıp gereksiz olanlarının atılması ile ilgili hazırlık süreci. 

3) verilerin dönüştürülmes i(data transformation): verilerin türlerini ya da dağılımını değiştirmek için kullanılan yöntemlerdir. makine öğrenmesinde sıralı ve kategorik veriler doğrudan kullanılamamaktadır. bunların algoritmalara sokulmadan önce sayısal biçime dönüştürülmesi gerekir. ayrık hale getirme. one hot encoding --> bu yöntemde kategorik veri birden fazla sütunu olan ikili (binary) veri haline dönüştürülmektedir. skala farklılıklarını ortadan kaldırmak için ölçekleme (scaling) işlemleri yapılabilir.

4) özellik mühendisliği (feature engineering): var olan sütunlardan yeni sütunlar oluşturma ile ilgili yöntemlere denilmetkedir. örneğin bir üstunda varlıklara ilişkin tarih bilgileri olabilir. biz de bu tarih bilgilerinden hareketle o tarihlerin hafta sonu olup olmadığını belirten yeni bir kategorik sütunu tabloya ekleyebiliriz. 

5) boyutsal özellik indirgemesi (dimensionalty feature reduction): çok boyutlu uzay. "çok sayıda sütundan bu sütunları temsil edebilecek daha az sayıda sütun" oluşturulması sürecidir. "temel bileşenler analizi (principal components analysis)"
