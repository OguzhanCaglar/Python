### Olaslk Kavram

--> bir olgunun gerekleme beklentisi. tura gelme olasl 0.5 - %50. "greli sklk (relative frequency)." 

tura gelme olasl / n --> 0.5. 1000000 atldnda yaz ile tura gelme says ayn olmaz. ancak tura gelme oran gitgide 0.5'e yaknsar. o halde bir bir parann atlmas durumunda tura gelme olaslnn 0.5 olmas demekle aslnda bir limit durumunu kastetmik olmaktayz



```python
import random

def head_tail(n):
    head = 0
    for _ in range(n):
        head += random.randint(0, 1) # tail = 0, head = 1
    
    return head / n
```

```python
head_tail(1)
```

```python
head_tail(10)
```

```python
head_tail(100)
```

```python
head_tail(10_000)
```

```python
head_tail(100_000)
```

```python
head_tail(1_000_000)
```

para atm says arttka tura gelme olasl da 0.5'e yaknsamaktadr. byk saylar yasas (law of large numbers).
